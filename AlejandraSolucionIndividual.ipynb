```python
import pandas as pd
import numpy as np
from IPython.display import display

# =====================================================================
# 1. FUNCIÓN EXIGIDA (ENFOQUE VECTORIZADO)
# =====================================================================
def year_from_title(df_movies):
    """
    Recibe el DataFrame de películas, extrae el año usando división de cadenas
    y limpia la columna 'title' sin usar expresiones regulares.
    """
    df = df_movies.copy()
    
    # limpieza básica inicial de espacios
    df['title'] = df['title'].astype(str).str.strip()
    
    # 1. Comprobar de forma vectorizada qué títulos terminan en ')'
    termina_en_parentesis = df['title'].str.endswith(')')
    
    # 2. Extraer el año rebanando los últimos caracteres fijos del final real
    # En lugar de -5:-1 que puede bailar, recortamos sabiendo que el paréntesis ocupa el lugar -1
    potential_years = df['title'].str[-5:-1]
    
    # 3. Forzar la conversión a numérico con pd.to_numeric
    df['year'] = pd.to_numeric(potential_years, errors="coerce").astype('Int64')
    
    # Validar que cumpla ambas condiciones: terminar en ')' y tener un número de 4 dígitos válido
    has_valid_year = termina_en_parentesis & df['year'].notna()
    
    # 4. Editar 'title' eliminando los últimos 7 caracteres " (YYYY)" solo a las válidas
    df.loc[has_valid_year, 'title'] = df.loc[has_valid_year, 'title'].str[:-7].str.strip()
    
    # Las que no tengan un año válido al final, se marcan explícitamente con pd.NA
    df.loc[~has_valid_year, 'year'] = pd.NA
    
    return df


# =====================================================================
# 2. EJECUCIÓN Y COMPROBACIONES EN PANTALLA
# =====================================================================
# Ejecutar la función sobre el dataset original
movies_clean = year_from_title(movies)

print("--- AUDITORÍA DE RESULTADOS (MÉTODO STR VECTORIZADO) --- \n")

no_year_movies = movies_clean[movies_clean['year'].isna()]
total_no_year = len(no_year_movies)

print(f"TOTAL DE PELÍCULAS SIN AÑO RECONOCIBLE: {total_no_year} (de un total de {len(movies_clean)})")
print("-" * 60)

print("Muestra de títulos detectados como CASOS LÍMITE:")
if total_no_year > 0:
    display(no_year_movies[['movieId', 'title']].head(10))

print("\n" + "="*70 + "\n")

print("VISTA PREVIA DEL DATAFRAME FINAL ('movies_clean')")
display(movies_clean.head(10))
```

```python
# --- ALEJANDRA ---
# --- Intento fallido con def year_from_title(title). Da error y no limpia correctamente el título, por lo que se opta por la versión vectorizada anterior ---
# He intentado seguir el title como argumento de la función, pero no funciona correctamente. Por ello, se opta por la versión vectorizada anterior.

def year_from_title(title):
    """
    Recibe un string de título, extrae el año entre paréntesis si sigue el patrón,
    y devuelve una tupla con (título_limpio, año) o (título_original, NaN).
    """
    if not isinstance(title, str):
        return title, np.nan
    
    title = title.strip()
    
    # Comprobar si termina en ')' y tiene un '(' 7 caracteres atrás
    if title.endswith(')') and len(title) >= 7 and title[-7] == '(':
        potential_year = title[-6:-1]
        
        # Validar que sean 4 dígitos numéricos
        if potential_year.isdigit() and len(potential_year) == 4:
            clean_title = title[:-7].strip()
            return clean_title, potential_year
        
    return title, np.nan

# =========================================================
# CREACIÓN DE MOVIES_CLEAN (La celda clave)
# =========================================================
movies_clean = movies.copy()

# Aplicar la función fila por fila y la expandimos en dos columnas
temp_cols = movies_clean['title'].apply(year_from_title).apply(pd.Series)

# Sobrescribir el título con la versión limpia (sin año)
movies_clean['title'] = temp_cols[0]

# Conversión de tipo numérico con enteros extensibles <NA>
movies_clean['year'] = pd.to_numeric(temp_cols[1], errors="coerce").astype('Int64')

# =====================================================================
# 3. COMPROBACIONES Y AUDITORÍA DE CASOS LÍMITE (REQUISITO CHECKLIST)
# =====================================================================
print("🔍 --- AUDITORÍA DE RESULTADOS Y CASOS LÍMITE --- \n")

# Filtrar las películas que se quedaron con año NaN (no reconocible
no_year_movies = movies_clean[movies_clean['year'].isna()]
total_no_year = len(no_year_movies)

# REQUISITO: Contar y mostrar cuántas películas no tienen año reconocible
print(f" TOTAL DE PELÍCULAS SIN AÑO RECONOCIBLE: {total_no_year} (de un total de {len(movies_clean)})")
print("-" * 60)

# REQUISITO: Mostrar muestras de los casos límite que no tenían patrón (YYYY)
print("Muestra de títulos detectados como CASOS LÍMITE (sin año válido al final):")
if total_no_year > 0:
    display(no_year_movies[['movieId', 'title']].head(10))
else:
    print("¡Ninguno! Todas las películas tenían un año válido.")

print("\n" + "="*70 + "\n")


# =====================================================================
# 4. VISUALIZACIÓN DE LA TABLA RESULTANTE (REQUISITO GENERAL)
# =====================================================================
print("📊 VISTA PREVIA DEL DATAFRAME FINAL ('movies_clean')")
print("Comprueba que el título NO tiene el año y que la columna 'year' está rellena:")
print("-" * 60)
display(movies_clean.head(10))
```

### 2. Unificación de datos (*merge* / *join*)

- Entender las **claves** entre tablas (p. ej. `movieId` une `movies`, `ratings`, `tags` y `links`).
- Construir un esquema unificado: al menos una tabla **película–usuario–rating** (ratings enriquecida con título y géneros) y otra con **película–tags** si procede.
- Usar `merge` (u operaciones equivalentes) con criterio claro: tipo de unión (*inner* / *left*), duplicados generados y cómo se resuelven.
- Imprimir las 10 primeras filas de las tablas resultantes.
- Dejar documentado qué filas se pierden o se multiplican al unir y **por qué** es aceptable en vuestro caso de uso.

```python
# --- ALEJANDRA ---
import pandas as pd
from IPython.display import display

# =========================================================
# 1. IDENTIFICACIÓN DE CLAVES (Análisis previo)
# =========================================================
# La clave principal de unión entre los conjuntos de datos es 'movieId'.
# - En 'movies_clean' actúa como Clave Primaria (ID único por película).
# - En 'ratings' y 'tags' actúa como Clave Foránea (puede repetirse).


# =========================================================
# 2. CONSTRUCCIÓN DE LA TABLA: PELÍCULA–USUARIO–RATING
# =========================================================
# REQUISITO: Ratings enriquecida con título y géneros.
# Criterio: LEFT JOIN desde 'ratings' para no perder ningún voto,
# incorporando la información del catálogo de películas de forma segura.

pelicula_usuario_rating = pd.merge(
    ratings, 
    movies_clean, 
    on='movieId', 
    how='left'
)


# =========================================================
# 3. CONSTRUCCIÓN DE LA TABLA: PELÍCULA–TAGS
# =========================================================
# REQUISITO: Tabla película-tags para análisis conceptual.
# Criterio: INNER JOIN porque solo interesan las películas 
# que efectivamente tienen etiquetas asignadas por los usuarios.

pelicula_tags = pd.merge(
    movies_clean, 
    tags, 
    on='movieId', 
    how='inner'
)


# =========================================================
# 4. IMPRESIÓN DE RESULTADOS (REQUISITO: 10 primeras filas)
# =========================================================

print("TABLA 1: película_usuario_rating (Primeras 10 filas)")
print(f"Dimensiones del DataFrame: {pelicula_usuario_rating.shape}")
display(pelicula_usuario_rating.head(10))

print("\n" + "="*60 + "\n")

print("TABLA 2: pelicula_tags (Primeras 10 filas)")
print(f"Dimensiones del DataFrame: {pelicula_tags.shape}")
display(pelicula_tags.head(10))
```

### 🔍 Justificación del Esquema Unificado y Gestión de Registros

#### 1. Tabla Película–Usuario–Rating (`how='left'`)
* **Tipo de Join:** Se ha aplicado un **`left join`** tomando la tabla `ratings` como eje de la izquierda.
* **Duplicados y Filas:** No se generan filas duplicadas de forma accidental debido a que `movieId` es una clave primaria única en la tabla de películas. La granularidad final se mantiene estrictamente a **nivel de interacción (voto)**.
* **Por qué es aceptable:** Este criterio garantiza que no perdamos ningún voto emitido por los usuarios bajo ninguna circunstancia. Si existiera alguna desincronización en el catálogo y una película votada no tuviera correspondencia de título en `movies_clean`, los campos de texto se rellenarían con `NaN`, pero el registro estadístico del voto se preservaría intacto, lo cual es prioritario para el negocio.

#### 2. Tabla Película–Tags (`how='inner'`)
* **Tipo de Join:** Se ha aplicado un **`inner join`** para cruzar ambas colecciones.
* **Duplicados y Filas:** En este merge, las filas de la tabla de películas **se multiplican legítimamente**. Si una película ha recibido 20 etiquetas distintas por parte de la comunidad, la información base de dicha película se repetirá 20 veces. La granularidad cambia a **nivel detalle de película-tag**.
* **Por qué es aceptable:** Es el comportamiento correcto y necesario para realizar análisis semánticos posteriores (como nubes de palabras o segmentación por etiquetas). Las películas que no han recibido ningún tag se descartan por completo de este DataFrame específico, ya que no aportan valor a un análisis que se enfoca puramente en el etiquetado conceptual.

- Usar `groupby` (por usuario, por película o por género) con al menos una agregación multi-columna (`agg`).
- Comparar dos segmentos (p. ej. usuarios con muchas valoraciones vs. pocos; o por **década** usando la columna `year` del apartado 3) con tablas resumen.

```python
# --- ALEJANDRA ---
print("--- INICIO DEL APARTADO 4: AGREGACIONES Y SEGMENTACIÓN --- \n")

# =====================================================================
# REQUISITO 1: GROUPBY CON AGREGACIÓN MULTI-COLUMNA (AGG)
# =====================================================================
# Agrupación por PELÍCULA para calcular métricas simultáneas de interés:
# Nota media, número total de votos y variabilidad (desviación estándar) de las notas.

# Primero necesitamos tener el dataset combinado de películas y ratings
# (Asegurar que 'movies_clean' y 'ratings' están cargados y ejecutados en el entorno)
if 'movies_clean' in locals() and 'ratings' in locals():
    movie_ratings = pd.merge(ratings, movies_clean, on='movieId', how='inner')
else:
    # Por si acaso tus variables se llaman diferente en el cuaderno colectivo:
    movie_ratings = pd.merge(ratings, movies_clean, on='movieId', how='inner')

print("1. Agregación Multi-columna por Película:")
print("Calculando media, conteo y desviación estándar de las valoraciones...")
print("-" * 60)

movie_summary = movie_ratings.groupby('title').agg({
    'rating': ['mean', 'count', 'std']
})

# Aplanamos los nombres de las columnas para que sea fácil trabajar con ellas
movie_summary.columns = ['nota_media', 'total_votos', 'desviacion_estandar']
movie_summary = movie_summary.reset_index()

# Mostramos las 10 películas con más votos para verificar que el cálculo es correcto
display(movie_summary.sort_values(by='total_votos', ascending=False).head(10))
print("\n" + "="*70 + "\n")


# =====================================================================
# REQUISITO 2: COMPARACIÓN DE DOS SEGMENTOS POR DÉCADA (TABLA RESUMEN)
# =====================================================================
print("2. Comparación de Segmentos por Década Histórica:")
print("Clasificando las películas en su década correspondiente...")
print("-" * 60)

# Filtro de registros que no tienen año para que no rompan la segmentación matemática
df_segmented = movie_ratings[movie_ratings['year'].notna()].copy()

# Creamos la columna 'decada' dividiendo el año de forma entera
df_segmented['decada'] = (df_segmented['year'] // 10) * 10
df_segmented['decada'] = df_segmented['decada'].astype(str) + 's'

# Agrupamos por década para construir la tabla resumen comparativa
decadas_summary = df_segmented.groupby('decada').agg({
    'movieId': 'nunique',     # Cuántas películas únicas se estrenaron en esa década
    'rating': ['mean', 'count'] # Nota media de la década y cuántos votos acumuló
})

# Aplanamos columnas del MultiIndex resultante
decadas_summary.columns = ['peliculas_unicas', 'nota_media_decada', 'volumen_total_votos']
decadas_summary = decadas_summary.reset_index()

# Ordenamos cronológicamente para que la tabla resumen sea legible
decadas_summary = decadas_summary.sort_values(by='decada')

# Mostramos la tabla comparativa final
display(decadas_summary)
```
