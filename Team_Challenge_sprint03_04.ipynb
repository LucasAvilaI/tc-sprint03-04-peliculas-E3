### 2. Columna `year` desde el título

- En el dataset de películas. el año de estreno suele aparecer **entre paréntesis al final** de `title`, p. ej. `Batman (1989)`.
- Implementad una función (`year_from_title`) que devuelva un entero de cuatro cifras o valor ausente (`NaN` / `<NA>`) si el título no sigue ese patrón.
- Añadid la columna **`year`** a `movies` como numérico (`pd.to_numeric(..., errors="coerce")`).
- Editar el campo `title` para que no contenga el año de estreno.
- Contad cuántas películas **no** tienen año reconocible y mostrad **una pequeña muestra** de sus títulos (casos límite).

```python
# --- RAUL ---
# --- Apartado 2: columna `year` ---

def year_from_title(titulo: str) -> int:
    try:
        # Por python, cualquier cosa que no sea un entero perfecto
        # positivo o negativo dará error.
        # Hago strip para quitar espacios al final y cosas raras.
        year = int(titulo.strip()[-5:-1])
        if year < 0: raise ValueError
        return year
    except ValueError:
        return pd.NA
    
df_movies["year"] = df_movies["title"].apply(year_from_title)

# RAUL: 
# Con la llamda anterior estaría todo cubierto
# Si me obligan a usar pd.to_numeric habría que hacer lo siguiente
# pero no le encuentro sentido. Además visualmente saca decimales.
# df_movies["year"] = round(pd.to_numeric(df_movies["year"], errors="coerce"), 0)

df_no_year = df_movies[df_movies["year"].isnull()]
print(f"El número de películas sin año son: {len(df_no_year)}")

# RAUL: 
# Esto se puede hacer con df_no_year["title"], pero ya no
# es un dataframe y al lanzar display se ve raro. 
display(df_no_year.loc[:,["title"]])

```

```python
# --- ALEJANDRA ---
import pandas as pd
from IPython.display import display

# =====================================================================
# 1. FUNCIÓN EXIGIDA (ENFOQUE VECTORIZADO)
# =====================================================================
def year_from_title(df_movies):
    """
    Recibe el DataFrame de películas, extrae el año usando división de cadenas
    y limpia la columna 'title' sin usar expresiones regulares.
    """
    df = df_movies.copy()
    
    # limpieza básica inicial de espacios
    df['title'] = df['title'].astype(str).str.strip()
    
    # 1. Comprobar de forma vectorizada qué títulos terminan en ')'
    termina_en_parentesis = df['title'].str.endswith(')')
    
    # 2. Extraer el año rebanando los últimos caracteres fijos del final real
    # En lugar de -5:-1 que puede bailar, recortamos sabiendo que el paréntesis ocupa el lugar -1
    potential_years = df['title'].str[-5:-1]
    
    # 3. Forzar la conversión a numérico con pd.to_numeric
    df['year'] = pd.to_numeric(potential_years, errors="coerce").astype('Int64')
    
    # Validar que cumpla ambas condiciones: terminar en ')' y tener un número de 4 dígitos válido
    has_valid_year = termina_en_parentesis & df['year'].notna()
    
    # 4. Editar 'title' eliminando los últimos 7 caracteres " (YYYY)" solo a las válidas
    df.loc[has_valid_year, 'title'] = df.loc[has_valid_year, 'title'].str[:-7].str.strip()
    
    # Las que no tengan un año válido al final, se marcan explícitamente con pd.NA
    df.loc[~has_valid_year, 'year'] = pd.NA
    
    return df


# =====================================================================
# 2. EJECUCIÓN Y COMPROBACIONES EN PANTALLA
# =====================================================================
# Ejecutar la función sobre el dataset original
movies_clean = year_from_title(movies)

print("--- AUDITORÍA DE RESULTADOS (MÉTODO STR VECTORIZADO) --- \n")

no_year_movies = movies_clean[movies_clean['year'].isna()]
total_no_year = len(no_year_movies)

print(f"TOTAL DE PELÍCULAS SIN AÑO RECONOCIBLE: {total_no_year} (de un total de {len(movies_clean)})")
print("-" * 60)

print("Muestra de títulos detectados como CASOS LÍMITE:")
if total_no_year > 0:
    display(no_year_movies[['movieId', 'title']].head(10))

print("\n" + "="*70 + "\n")

print("VISTA PREVIA DEL DATAFRAME FINAL ('movies_clean')")
display(movies_clean.head(10))

movies_clean.to_csv("data/df_ale/movies_clean_checkpoint.csv", index=False)
print("Checkpoint de data set de películas limpio 'movies_clean' guardado en 'data/df_ale/movies_clean_checkpoint.csv'")
```

### 4. Agregaciones y segmentación

- Usar `groupby` (por usuario, por película o por género) con al menos una agregación multi-columna (`agg`).
- Comparar dos segmentos (p. ej. usuarios con muchas valoraciones vs. pocos; o por **década** usando la columna `year` del apartado 3) con tablas resumen.

```python
# --- Apartado 4: Agregaciones ---
```

```python
# --- RAUL ---
# --- Apartado 4: Agregaciones ---

# RAUL: Devuelve por película la nota media, la máxima y la mínima.
df_notas = df_mov_x_rat.groupby(by=["movieId", "title"])["rating"].agg(["mean", "max", "min"]).round({"mean": 2})
#display(df_notas)

# RAUL: Vamos a contar cuantas películas son estan aprobadas o suspendidas. 

df_aprobadas = df_notas[df_notas["mean"] >= 2.5]
df_aprobadas["Nota"] = "Aprobada"

df_suspendidas = df_notas[df_notas["mean"] < 2.5]
df_suspendidas["Nota"] = "Suspendida"

display(((pd.concat([df_aprobadas, df_suspendidas])).groupby("Nota").count())[["mean"]].rename(columns={"mean": "cantidad"}))
```

```python
# --- ALEJANDRA ---
print("--- INICIO DEL APARTADO 4: AGREGACIONES Y SEGMENTACIÓN --- \n")

# =====================================================================
# REQUISITO 1: GROUPBY CON AGREGACIÓN MULTI-COLUMNA (AGG)
# =====================================================================
# Agrupación por PELÍCULA para calcular métricas simultáneas de interés:
# Nota media, número total de votos y variabilidad (desviación estándar) de las notas.

# Primero necesitamos tener el dataset combinado de películas y ratings
# (Asegurar que 'movies_clean' y 'ratings' están cargados y ejecutados en el entorno)
if 'movies_clean' in locals() and 'ratings' in locals():
    movie_ratings = pd.merge(ratings, movies_clean, on='movieId', how='inner')
else:
    # Se guarda la variable 'movie_ratings' para evitar errores
    # si los nombres de los DataFrames son distintos:
    movie_ratings = pd.merge(ratings, movies_clean, on='movieId', how='inner')

# Se guardan los resultados del csv de la variable 'movie_ratings' para persistencia
# y evitar errores si se ejecuta en otro entorno
movie_ratings.to_csv("data/df_ale/ratings_movies_checkpoint.csv", index=False)
print("Checkpoint de la Parte 1 guardado en 'data/ratings_movies_checkpoint.csv'")

print("1. Agregación Multi-columna por Película:")
print("Calculando media, conteo y desviación estándar de las valoraciones...")
print("-" * 60)

movie_summary = movie_ratings.groupby('title').agg({
    'rating': ['mean', 'count', 'std']
})

# Aplanamos los nombres de las columnas para que sea fácil trabajar con ellas
movie_summary.columns = ['nota_media', 'total_votos', 'desviacion_estandar']
movie_summary = movie_summary.reset_index()

# Mostramos las 10 películas con más votos para verificar que el cálculo es correcto
display(movie_summary.sort_values(by='total_votos', ascending=False).head(10))
print("\n" + "="*70 + "\n")


# =====================================================================
# REQUISITO 2: COMPARACIÓN DE DOS SEGMENTOS POR DÉCADA (TABLA RESUMEN)
# =====================================================================
print("2. Comparación de Segmentos por Década Histórica:")
print("Clasificando las películas en su década correspondiente...")
print("-" * 60)

# Filtro de registros que no tienen año para que no rompan la segmentación matemática
df_segmented = movie_ratings[movie_ratings['year'].notna()].copy()

# Creamos la columna 'decada' dividiendo el año de forma entera
df_segmented['decada'] = (df_segmented['year'] // 10) * 10
df_segmented['decada'] = df_segmented['decada'].astype(str) + 's'

# Agrupamos por década para construir la tabla resumen comparativa
decadas_summary = df_segmented.groupby('decada').agg({
    'movieId': 'nunique',     # Cuántas películas únicas se estrenaron en esa década
    'rating': ['mean', 'count'] # Nota media de la década y cuántos votos acumuló
})

# Aplanamos columnas del MultiIndex resultante
decadas_summary.columns = ['peliculas_unicas', 'nota_media_decada', 'volumen_total_votos']
decadas_summary = decadas_summary.reset_index()

# Ordenamos cronológicamente para que la tabla resumen sea legible
decadas_summary = decadas_summary.sort_values(by='decada')

# Mostramos la tabla comparativa final
display(decadas_summary)
```

```python
# --- ALEJANDRA ---
# =====================================================================
# PARTE 3: SINOPSIS EN ESPAÑOL CON GEMINI (VERSIÓN FINAL BLINDADA)
# =====================================================================

import os
import getpass
import time
import json
import urllib.request
import urllib.parse
import pandas as pd
from google import genai

# =====================================================================
# 1. SEGURIDAD: PROTECCIÓN DE CREDENCIALES
# =====================================================================
print("[SEGURIDAD] Ejecutando auditoría de credenciales...")

# Escaneo preventivo de archivos expuestos
for archivo in ['.env', 'config.env', 'settings.env', '.secrets']:
    if os.path.exists(archivo):
        print(f"[ALERTA]: Se detectó '{archivo}'. ¡Añádelo a tu .gitignore!")

# Entrada manual y segura con getpass para no dejar rastro en el sistema
raw_key = getpass.getpass("Introduce tu GEMINI_API_KEY de Google AI Studio: ")
clean_key = raw_key.strip() if raw_key else ""

if not clean_key or len(clean_key) < 10:
    raise ValueError("Error: API Key inválida.")

# Inyección temporal para inicializar el cliente
os.environ["GEMINI_API_KEY"] = clean_key
client = genai.Client()


# =====================================================================
# 2. FUNCIÓN DE PROCESAMIENTO CON FALLBACK MULTINIVEL
# =====================================================================
def summarize_overview_es(overview, title=""):
    if pd.isna(overview) or str(overview).strip() == "" or overview == "No overview available.":
        return ""
    
    prompt = f"Traduce y resume en español la sinopsis de '{title}' en máximo 2 frases: {overview}"
    
    # Capa A: Gemini 2.5-flash (Principal)
    try:
        response = client.models.generate_content(model='gemini-2.5-flash', contents=prompt)
        return response.text.strip()
    except Exception as e:
        print(f"[FALLO CAPA A] {e}")
        
        # Capa B: Gemini 1.5-flash (Backup IA)
        try:
            response = client.models.generate_content(model='gemini-1.5-flash', contents=prompt)
            return response.text.strip()
        except:
            # Capa C: Traductor Google (Zero-Cost / Online)
            try:
                url = "https://translate.googleapis.com/translate_a/single?client=gtx&sl=en&tl=es&dt=t&q=" + urllib.parse.quote(str(overview))
                req = urllib.request.Request(url, headers={'User-Agent': 'Mozilla/5.0'})
                with urllib.request.urlopen(req) as resp:
                    data = json.loads(resp.read().decode('utf-8'))
                    return "".join([s[0] for s in data[0] if s[0]])
            except:
                return f"[Error en traducción] original: {str(overview)[:40]}..."


# =====================================================================
# 3. BUCLE DE EJECUCIÓN: VENTANA COMPLETA CON BREAKS
# =====================================================================
if 'movies10' not in globals():
    movies10 = pd.read_csv("data/df_ale/movies10_with_overviews.csv")

movies10_processed = movies10.copy()
resumenes_es = []
contador_llamadas = 0
MAX_LLAMADAS = 15  # Límite financiero absoluto

print("\n Iniciando procesamiento completo (10 películas)...")

for idx, row in movies10_processed.iterrows():
    # FILTRO FINANCIERO: Si llegamos al límite, paramos
    if contador_llamadas >= MAX_LLAMADAS:
        print(f"[FINANCIAL SHUTDOWN] Límite de {MAX_LLAMADAS} alcanzado.")
        break
        
    print(f"Procesando ({idx+1}/{len(movies10_processed)}): {row.get('title', '')}...")
    
    resumen = summarize_overview_es(row.get('overview', ''), title=row.get('title', ''))
    resumenes_es.append(resumen)
    
    # Contamos solo las llamadas que no sean fallos técnicos o bloqueos
    if resumen != "" and not resumen.startswith("["):
        contador_llamadas += 1
        
    # BREAKS ESTRATÉGICOS: Pausa tras cada 5 ítems para estabilidad
    if (idx + 1) % 5 == 0:
        print("[BREAK] Pausa de seguridad de 10 segundos...")
        time.sleep(10)
    else:
        time.sleep(3)

# Completar con vacíos si el bucle paró antes de tiempo
while len(resumenes_es) < len(movies10_processed):
    resumenes_es.append("[No procesado]")

movies10_processed["overview_es"] = resumenes_es


# =====================================================================
# 4. LIMPIEZA FINAL (DESTRUCCIÓN DE CLAVE)
# =====================================================================
if "GEMINI_API_KEY" in os.environ:
    del os.environ["GEMINI_API_KEY"]
print("\n [SEGURIDAD] Credenciales destruidas. Entorno limpio.")


# =====================================================================
# 5. VISUALIZACIÓN
# =====================================================================
print("\n🔍 === RESULTADOS OBLIGATORIOS (VISTA DE 3 PELÍCULAS) ===")
display(movies10_processed[["title", "overview_es"]].head(3))
```

```python
# --- ALEJANDRA ---
# =====================================================================
# PARTE OPCIONAL A: ENRIQUECIMIENTO CON ELIMINACIÓN AUTOMÁTICA DE CLAVE
# =====================================================================

import os
import json
import time
import getpass
import pandas as pd
from google import genai
from google.genai import types

# =====================================================================
# 1. CONTROL ESTRICTO DE CREDENCIALES (AISLAMIENTO Y LIMPIEZA)
# =====================================================================
print("[SEGURIDAD] Configurando entorno aislado para rama Alejandra...")

# Solicitamos la clave en una variable temporal
raw_key = getpass.getpass("Introduce tu GEMINI_API_KEY (Entrada oculta y obligatoria): ")

# .strip() elimina CUALQUIER espacio en blanco o salto de línea accidental al pegar
clean_key = raw_key.strip() if raw_key else ""

if not clean_key or len(clean_key) < 10:
    raise ValueError("Error: La API Key está vacía o es demasiado corta.")

# Asignamos temporalmente al entorno para que el inicializador nativo la lea
os.environ["GEMINI_API_KEY"] = clean_key

# Inicializamos el cliente oficial
client = genai.Client()


# =====================================================================
# 2. FUNCIÓN DE ENRIQUECIMIENTO RESILIENTE (CON MODELO 2.5 NATIVO)
# =====================================================================
def enrich_catalog_fields(overview, title="", genres="", year=None):
    """
    Enriquece los metadatos de una película generando un JSON estructurado nativo con Gemini 2.5.
    """
    if pd.isna(overview) or str(overview).strip() == "" or overview == "No overview available.":
        return None

    prompt = f"""
    Analiza la sinopsis y metadatos de la película:
    Título: {title}
    Año: {year}
    Géneros: {genres}
    Sinopsis: {overview}
    """
    
    config = types.GenerateContentConfig(
        response_mime_type="application/json",
        response_schema={
            "type": "OBJECT",
            "properties": {
                "pitch_es": {"type": "STRING", "description": "Texto promocional de cartelera en español, máx 280 caracteres."},
                "edad_sugerida": {"type": "STRING", "enum": ["TP", "+7", "+12", "+16", "+18"]},
                "temas": {"type": "STRING", "description": "Exactamente 3 temas en español separados por comas."}
            },
            "required": ["pitch_es", "edad_sugerida", "temas"]
        },
        temperature=0.2
    )
    
    try:
        response = client.models.generate_content(
            model='gemini-2.5-flash',
            contents=prompt,
            config=config
        )
        
        data = json.loads(response.text)
        return {
            "pitch_es": data.get("pitch_es", "")[:280],
            "edad_sugerida": data.get("edad_sugerida", "TP"),
            "temas": data.get("temas", "")
        }
    except Exception as e:
        print(f"Error procesando la película '{title}': {e}")
        return None


# =====================================================================
# 3. EJECUCIÓN DEL BUCLE CON CONTROL DE COSTES INTERNO
# =====================================================================
movies10_enriched = movies10.copy()
pitches, edades, lista_temas = [], [], []

# CONTROL FINANCIERO LOCAL (Máximo 15 llamadas en esta ejecución)
MAX_LLAMADAS_PERMITIDAS = 15  
contador_llamadas = 0

print("\n Iniciando enriquecimiento con esquema estructurado JSON...")

for idx, row in movies10_enriched.iterrows():
    context_overview = row.get('overview_es', row.get('overview', ''))
    titulo = row.get('title', '')
    
    if contador_llamadas >= MAX_LLAMADAS_PERMITIDAS:
        print(f"[FRENO DE MANO] Límite local de {MAX_LLAMADAS_PERMITIDAS} llamadas alcanzado.")
        pitches.append(None); edades.append(None); lista_temas.append(None)
        continue
        
    print(f"Analizando ({idx+1}/{len(movies10_enriched)}): {titulo}...")
    
    res = enrich_catalog_fields(
        context_overview, 
        title=titulo, 
        genres=row.get('genres', ''), 
        year=row.get('year', None)
    )
    
    if res:
        pitches.append(res["pitch_es"])
        edades.append(res["edad_sugerida"])
        lista_temas.append(res["temas"])
        contador_llamadas += 1
    else:
        pitches.append(None); edades.append(None); lista_temas.append(None)
        
    if idx < len(movies10_enriched) - 1:
        time.sleep(2)

# Asignación de los vectores resultantes al DataFrame final
movies10_enriched["pitch_es"] = pitches
movies10_enriched["edad_sugerida"] = edades
movies10_enriched["temas"] = lista_temas


# =====================================================================
# 4. LIMPIEZA POST-EJECUCIÓN (PROTECCIÓN TOTAL ANTE COMPAÑEROS)
# =====================================================================
# Borramos de forma fulminante la clave del entorno para que nadie más pueda usarla
if "GEMINI_API_KEY" in os.environ:
    del os.environ["GEMINI_API_KEY"]
print("\n [SEGURIDAD] Credenciales destruidas con éxito de la sesión de ejecución.")


# =====================================================================
# 5. VISUALIZACIÓN DE CONTROL DEL DATAFRAME
# =====================================================================
print("=== RESULTADOS FINALES ENRIQUECIDOS (VISTA DE 3 PELÍCULAS) ===")
display(movies10_enriched[["title", "pitch_es", "edad_sugerida", "temas"]].head(3))
```
