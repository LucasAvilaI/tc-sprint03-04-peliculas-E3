### 2. Columna `year` desde el título

- En el dataset de películas. el año de estreno suele aparecer **entre paréntesis al final** de `title`, p. ej. `Batman (1989)`.
- Implementad una función (`year_from_title`) que devuelva un entero de cuatro cifras o valor ausente (`NaN` / `<NA>`) si el título no sigue ese patrón.
- Añadid la columna **`year`** a `movies` como numérico (`pd.to_numeric(..., errors="coerce")`).
- Editar el campo `title` para que no contenga el año de estreno.
- Contad cuántas películas **no** tienen año reconocible y mostrad **una pequeña muestra** de sus títulos (casos límite).

```python
# --- RAUL ---
# --- Apartado 2: columna `year` ---

def year_from_title(titulo: str) -> int:
    try:
        # Por python, cualquier cosa que no sea un entero perfecto
        # positivo o negativo dará error.
        # Hago strip para quitar espacios al final y cosas raras.
        year = int(titulo.strip()[-5:-1])
        if year < 0: raise ValueError
        return year
    except ValueError:
        return pd.NA
    
df_movies["year"] = df_movies["title"].apply(year_from_title)

# RAUL: 
# Con la llamda anterior estaría todo cubierto
# Si me obligan a usar pd.to_numeric habría que hacer lo siguiente
# pero no le encuentro sentido. Además visualmente saca decimales.
# df_movies["year"] = round(pd.to_numeric(df_movies["year"], errors="coerce"), 0)

df_no_year = df_movies[df_movies["year"].isnull()]
print(f"El número de películas sin año son: {len(df_no_year)}")

# RAUL: 
# Esto se puede hacer con df_no_year["title"], pero ya no
# es un dataframe y al lanzar display se ve raro. 
display(df_no_year.loc[:,["title"]])

```

```python
# --- ALEJANDRA ---
import pandas as pd
from IPython.display import display

# =====================================================================
# 1. FUNCIÓN EXIGIDA (ENFOQUE VECTORIZADO)
# =====================================================================
def year_from_title(df_movies):
    """
    Recibe el DataFrame de películas, extrae el año usando división de cadenas
    y limpia la columna 'title' sin usar expresiones regulares.
    """
    df = df_movies.copy()
    
    # limpieza básica inicial de espacios
    df['title'] = df['title'].astype(str).str.strip()
    
    # 1. Comprobar de forma vectorizada qué títulos terminan en ')'
    termina_en_parentesis = df['title'].str.endswith(')')
    
    # 2. Extraer el año rebanando los últimos caracteres fijos del final real
    # En lugar de -5:-1 que puede bailar, recortamos sabiendo que el paréntesis ocupa el lugar -1
    potential_years = df['title'].str[-5:-1]
    
    # 3. Forzar la conversión a numérico con pd.to_numeric
    df['year'] = pd.to_numeric(potential_years, errors="coerce").astype('Int64')
    
    # Validar que cumpla ambas condiciones: terminar en ')' y tener un número de 4 dígitos válido
    has_valid_year = termina_en_parentesis & df['year'].notna()
    
    # 4. Editar 'title' eliminando los últimos 7 caracteres " (YYYY)" solo a las válidas
    df.loc[has_valid_year, 'title'] = df.loc[has_valid_year, 'title'].str[:-7].str.strip()
    
    # Las que no tengan un año válido al final, se marcan explícitamente con pd.NA
    df.loc[~has_valid_year, 'year'] = pd.NA
    
    return df


# =====================================================================
# 2. EJECUCIÓN Y COMPROBACIONES EN PANTALLA
# =====================================================================
# Ejecutar la función sobre el dataset original
movies_clean = year_from_title(movies)

print("--- AUDITORÍA DE RESULTADOS (MÉTODO STR VECTORIZADO) --- \n")

no_year_movies = movies_clean[movies_clean['year'].isna()]
total_no_year = len(no_year_movies)

print(f"TOTAL DE PELÍCULAS SIN AÑO RECONOCIBLE: {total_no_year} (de un total de {len(movies_clean)})")
print("-" * 60)

print("Muestra de títulos detectados como CASOS LÍMITE:")
if total_no_year > 0:
    display(no_year_movies[['movieId', 'title']].head(10))

print("\n" + "="*70 + "\n")

print("VISTA PREVIA DEL DATAFRAME FINAL ('movies_clean')")
display(movies_clean.head(10))

movies_clean.to_csv("data/df_ale/movies_clean_checkpoint.csv", index=False)
print("Checkpoint de data set de películas limpio 'movies_clean' guardado en 'data/df_ale/movies_clean_checkpoint.csv'")
```

### 4. Agregaciones y segmentación

- Usar `groupby` (por usuario, por película o por género) con al menos una agregación multi-columna (`agg`).
- Comparar dos segmentos (p. ej. usuarios con muchas valoraciones vs. pocos; o por **década** usando la columna `year` del apartado 3) con tablas resumen.

```python
# --- Apartado 4: Agregaciones ---
```

```python
# --- RAUL ---
# --- Apartado 4: Agregaciones ---

# RAUL: Devuelve por película la nota media, la máxima y la mínima.
df_notas = df_mov_x_rat.groupby(by=["movieId", "title"])["rating"].agg(["mean", "max", "min"]).round({"mean": 2})
#display(df_notas)

# RAUL: Vamos a contar cuantas películas son estan aprobadas o suspendidas. 

df_aprobadas = df_notas[df_notas["mean"] >= 2.5]
df_aprobadas["Nota"] = "Aprobada"

df_suspendidas = df_notas[df_notas["mean"] < 2.5]
df_suspendidas["Nota"] = "Suspendida"

display(((pd.concat([df_aprobadas, df_suspendidas])).groupby("Nota").count())[["mean"]].rename(columns={"mean": "cantidad"}))
```

```python
# --- ALEJANDRA ---
print("--- INICIO DEL APARTADO 4: AGREGACIONES Y SEGMENTACIÓN --- \n")

# =====================================================================
# REQUISITO 1: GROUPBY CON AGREGACIÓN MULTI-COLUMNA (AGG)
# =====================================================================
# Agrupación por PELÍCULA para calcular métricas simultáneas de interés:
# Nota media, número total de votos y variabilidad (desviación estándar) de las notas.

# Primero necesitamos tener el dataset combinado de películas y ratings
# (Asegurar que 'movies_clean' y 'ratings' están cargados y ejecutados en el entorno)
if 'movies_clean' in locals() and 'ratings' in locals():
    movie_ratings = pd.merge(ratings, movies_clean, on='movieId', how='inner')
else:
    # Se guarda la variable 'movie_ratings' para evitar errores
    # si los nombres de los DataFrames son distintos:
    movie_ratings = pd.merge(ratings, movies_clean, on='movieId', how='inner')

# Se guardan los resultados del csv de la variable 'movie_ratings' para persistencia
# y evitar errores si se ejecuta en otro entorno
movie_ratings.to_csv("data/df_ale/ratings_movies_checkpoint.csv", index=False)
print("Checkpoint de la Parte 1 guardado en 'data/ratings_movies_checkpoint.csv'")

print("1. Agregación Multi-columna por Película:")
print("Calculando media, conteo y desviación estándar de las valoraciones...")
print("-" * 60)

movie_summary = movie_ratings.groupby('title').agg({
    'rating': ['mean', 'count', 'std']
})

# Aplanamos los nombres de las columnas para que sea fácil trabajar con ellas
movie_summary.columns = ['nota_media', 'total_votos', 'desviacion_estandar']
movie_summary = movie_summary.reset_index()

# Mostramos las 10 películas con más votos para verificar que el cálculo es correcto
display(movie_summary.sort_values(by='total_votos', ascending=False).head(10))
print("\n" + "="*70 + "\n")


# =====================================================================
# REQUISITO 2: COMPARACIÓN DE DOS SEGMENTOS POR DÉCADA (TABLA RESUMEN)
# =====================================================================
print("2. Comparación de Segmentos por Década Histórica:")
print("Clasificando las películas en su década correspondiente...")
print("-" * 60)

# Filtro de registros que no tienen año para que no rompan la segmentación matemática
df_segmented = movie_ratings[movie_ratings['year'].notna()].copy()

# Creamos la columna 'decada' dividiendo el año de forma entera
df_segmented['decada'] = (df_segmented['year'] // 10) * 10
df_segmented['decada'] = df_segmented['decada'].astype(str) + 's'

# Agrupamos por década para construir la tabla resumen comparativa
decadas_summary = df_segmented.groupby('decada').agg({
    'movieId': 'nunique',     # Cuántas películas únicas se estrenaron en esa década
    'rating': ['mean', 'count'] # Nota media de la década y cuántos votos acumuló
})

# Aplanamos columnas del MultiIndex resultante
decadas_summary.columns = ['peliculas_unicas', 'nota_media_decada', 'volumen_total_votos']
decadas_summary = decadas_summary.reset_index()

# Ordenamos cronológicamente para que la tabla resumen sea legible
decadas_summary = decadas_summary.sort_values(by='decada')

# Mostramos la tabla comparativa final
display(decadas_summary)
```

```python
# --- ALEJANDRA ---
import json
import os
import time
from getpass import getpass
from typing import Literal
import pandas as pd
from google import genai
from google.genai import types
from pydantic import BaseModel, Field

if "GEMINI_API_KEY" not in os.environ:
    os.environ["GEMINI_API_KEY"] = getpass("Introduce tu Gemini API Key: ")

client = genai.Client()

# =====================================================================
# CONFIGURACIÓN DE SEGURIDAD PARA DATAFRAMES DE ORIGEN
# =====================================================================
if 'ratings_movies' not in locals() and 'ratings_movies' not in globals():
    if 'ratings' in locals() or 'ratings' in globals():
        ratings_movies = ratings
    elif 'df_ratings' in locals() or 'df_ratings' in globals():
        ratings_movies = df_ratings
    else:
        raise NameError("No se encontró el DataFrame de valoraciones. Por favor, asigna tu DataFrame a la variable 'ratings_movies' antes de continuar.")
    
# Para asegurar persistencia de variables y resultados, se recomienda guardar 'ratings_movies'
# en un CSV local en carpeta data si se desea reutilizar sin volver a llamar a la API.
ratings_movies.to_csv("data/ratings_movies_checkpoint.csv", index=False)

# =====================================================================
# PARTE 4 - SECCIÓN A: Catálogo ampliado en movies10 (Llamada Optimizada)
# =====================================================================

class MovieItem(BaseModel):
    movieId: int = Field(description="El ID numérico correspondiente a la película procesada.")
    pitch_es: str = Field(description="Texto de cartelera en español, máximo 280 caracteres.")
    edad_sugerida: Literal['TP', '+7', '+12', '+16', '+18'] = Field(description="Clasificación de edad sugerida.")
    temas: str = Field(description="Exactamente 3 temas clave en español, separados exclusivamente por comas.")

class BulkMovieEnrichment(BaseModel):
    peliculas: list[MovieItem] = Field(description="Lista que contiene el enriquecimiento de cada una de las películas enviadas.")

def enrich_catalog_bulk(df):
    input_data = []
    for idx, row in df.iterrows():
        input_data.append({
            "movieId": int(row['movieId']),
            "title": row.get('title', ""),
            "genres": row.get('genres', ""),
            "overview": row.get('overview', "")
        })
    
    prompt = f"""
    Analiza la lista de películas proporcionada en formato JSON. Para cada elemento, genera los campos requeridos ('pitch_es', 'edad_sugerida', 'temas') basándote estrictamente en su sinopsis ('overview') y sus géneros. Si una sinopsis está vacía, devuelve valores nulos para esa película.
    
    Catálogo a procesar:
    {json.dumps(input_data, ensure_ascii=False)}
    """
    
    try:
        response = client.models.generate_content(
            model='gemini-2.5-flash',
            contents=prompt,
            config=types.GenerateContentConfig(
                response_mime_type="application/json",
                response_schema=BulkMovieEnrichment,
                temperature=0.2
            ),
        )
        return json.loads(response.text)
    except Exception as e:
        print(f"Error en el procesamiento masivo: {e}")
        return {"peliculas": []}

df_enriched = movies10.copy()
print("Enriqueciendo el catálogo completo mediante llamada optimizada...")
result_bulk = enrich_catalog_bulk(df_enriched)

# Mapeo de los resultados devueltos al DataFrame de origen
if "peliculas" in result_bulk and result_bulk["peliculas"]:
    enrich_df = pd.DataFrame(result_bulk["peliculas"])
    df_enriched = pd.merge(df_enriched, enrich_df[['movieId', 'pitch_es', 'edad_sugerida', 'temas']], on='movieId', how='left')
else:
    df_enriched['pitch_es'] = None
    df_enriched['edad_sugerida'] = None
    df_enriched['temas'] = None

print("Muestra del catálogo ampliado (A):")
print(df_enriched[['title', 'pitch_es', 'edad_sugerida', 'temas']].head(3))

# Para asegurar persistencia de variables y resultados, se recomienda guardar 'df_enriched'
# en un CSV local en carpeta data si se desea reutilizar sin volver a llamar a la API.
df_enriched.to_csv("data/movies10_enriched.csv", index=False)

# Pausa de seguridad para mitigar el límite de peticiones por minuto (Rate Limits)
print("\nEsperando 15 segundos para liberar la cuota de la API antes de la recomendación...")
time.sleep(15)


# =====================================================================
# PARTE 4 - SECCIÓN B: Recomendador por género
# =====================================================================

class MovieRecommendation(BaseModel):
    movieId: int = Field(description="El ID numérico de la película recomendada que pertenece al catálogo candidato.")
    title: str = Field(description="El título exacto de la película recomendada.")
    justificacion_es: str = Field(description="Una única frase de justificación en español explicando por qué se recomienda basándose en los gustos.")

class RecommendationList(BaseModel):
    recomendaciones: list[MovieRecommendation] = Field(description="Lista que contiene exactamente 3 recomendaciones únicas de películas.")

def top_rated_by_genre(ratings_movies, genres, top_n=10, min_ratings=50):
    df_filtered = pd.merge(ratings_movies, movies10[['movieId', 'title', 'genres']], on='movieId', how='inner')
    
    if isinstance(genres, list):
        condition = df_filtered['genres'].apply(lambda x: any(g.lower() in str(x).lower() for g in genres))
        df_filtered = df_filtered[condition]
    else:
        df_filtered = df_filtered[df_filtered['genres'].str.contains(genres, case=False, na=False)]
        
    stats = df_filtered.groupby('movieId').agg(
        title=('title', 'first'),
        genres=('genres', 'first'),
        rating_mean=('rating', 'mean'),
        rating_count=('rating', 'count')
    ).reset_index()
    
    candidates = stats[stats['rating_count'] >= min_ratings]
    candidates = candidates.sort_values(by='rating_mean', ascending=False).head(top_n)
    return candidates

def recommend_movies(candidates, favorite_genres, n=3):
    if candidates.empty:
        return {"recomendaciones": [], "status": "No hay películas candidatas para procesar."}
        
    candidates_list = candidates[['movieId', 'title', 'genres', 'rating_mean']].to_dict(orient='records')
    
    prompt = f"""
    Actúa como un sistema experto de recomendación cinematográfica. 
    A continuación tienes un catálogo de películas candidatas con mejores puntuaciones:
    {json.dumps(candidates_list, ensure_ascii=False)}
    
    Los géneros favoritos del usuario son: {favorite_genres}
    
    Selecciona exactamente {n} películas de la lista de candidatos que mejor se adapten a los géneros favoritos. 
    Para cada una proporciona una breve justificación de una única frase en español. 
    No inventes películas ni uses títulos fuera de la lista de candidatos proporcionada.
    """
    
    try:
        response = client.models.generate_content(
            model='gemini-2.5-flash',
            contents=prompt,
            config=types.GenerateContentConfig(
                response_mime_type="application/json",
                response_schema=RecommendationList,
                temperature=0.3
            ),
        )
        return json.loads(response.text)
    except Exception as e:
        return {"error": str(e)}

generos_prueba = ["Action", "Sci-Fi"]
df_candidatas = top_rated_by_genre(ratings_movies, genres=generos_prueba, top_n=10, min_ratings=50)

print("\nPelículas candidatas filtradas (B.1):")
print(df_candidatas[['movieId', 'title', 'rating_mean', 'rating_count']])

resultado_recomendacion = recommend_movies(df_candidatas, favorite_genres=generos_prueba, n=3)

print("\nRecomendaciones del modelo Gemini (B.2):")
print(json.dumps(resultado_recomendacion, indent=4, ensure_ascii=False))
```
