- En el dataset de películas. el año de estreno suele aparecer **entre paréntesis al final** de `title`, p. ej. `Batman (1989)`.
- Implementad una función (`year_from_title`) que devuelva un entero de cuatro cifras o valor ausente (`NaN` / `<NA>`) si el título no sigue ese patrón.
- Añadid la columna **`year`** a `movies` como numérico (`pd.to_numeric(..., errors="coerce")`).
- Editar el campo `title` para que no contenga el año de estreno.

```python
# --- RAUL ---
# --- Apartado 2: columna `year` ---

def year_from_title(titulo: str) -> int:
    try:
        # Por python, cualquier cosa que no sea un entero perfecto
        # positivo o negativo dará error.
        # Hago strip para quitar espacios al final y cosas raras.
        year = int(titulo.strip()[-5:-1])
        if year < 0: raise ValueError
        return year
    except ValueError:
        return pd.NA
    
df_movies["year"] = df_movies["title"].apply(year_from_title)

# RAUL: 
# Con la llamda anterior estaría todo cubierto
# Si me obligan a usar pd.to_numeric habría que hacer lo siguiente
# pero no le encuentro sentido. Además visualmente saca decimales.
# df_movies["year"] = round(pd.to_numeric(df_movies["year"], errors="coerce"), 0)

df_no_year = df_movies[df_movies["year"].isnull()]
print(f"El número de películas sin año son: {len(df_no_year)}")

# RAUL: 
# Esto se puede hacer con df_no_year["title"], pero ya no
# es un dataframe y al lanzar display se ve raro. 
display(df_no_year.loc[:,["title"]])

```

```python
# --- ALEJANDRA ---
import pandas as pd
from IPython.display import display

# =====================================================================
# 1. FUNCIÓN EXIGIDA (ENFOQUE VECTORIZADO SIN REGEX)
# =====================================================================
def year_from_title(df_movies):
    """
    Recibe el DataFrame de películas, extrae el año usando división de cadenas
    y limpia la columna 'title' sin usar expresiones regulares.
    """
    df = df_movies.copy()
    
    # limpieza básica inicial de espacios
    df['title'] = df['title'].astype(str).str.strip()
    
    # 1. Comprobar de forma vectorizada qué títulos terminan en ')'
    termina_en_parentesis = df['title'].str.endswith(')')
    
    # 2. Extraer el año rebanando los últimos caracteres fijos del final real
    # En lugar de -5:-1 que puede bailar, recortamos sabiendo que el paréntesis ocupa el lugar -1
    potential_years = df['title'].str[-5:-1]
    
    # 3. Forzar la conversión a numérico con pd.to_numeric
    df['year'] = pd.to_numeric(potential_years, errors="coerce").astype('Int64')
    
    # Validar que cumpla ambas condiciones: terminar en ')' y tener un número de 4 dígitos válido
    has_valid_year = termina_en_parentesis & df['year'].notna()
    
    # 4. Editar 'title' eliminando los últimos 7 caracteres " (YYYY)" solo a las válidas
    df.loc[has_valid_year, 'title'] = df.loc[has_valid_year, 'title'].str[:-7].str.strip()
    
    # Las que no tengan un año válido al final, se marcan explícitamente con pd.NA
    df.loc[~has_valid_year, 'year'] = pd.NA
    
    return df


# =====================================================================
# 2. EJECUCIÓN Y COMPROBACIONES EN PANTALLA
# =====================================================================
# Ejecutar la función sobre el dataset original
movies_clean = year_from_title(movies)

print("--- AUDITORÍA DE RESULTADOS (MÉTODO STR VECTORIZADO) --- \n")

no_year_movies = movies_clean[movies_clean['year'].isna()]
total_no_year = len(no_year_movies)

print(f"TOTAL DE PELÍCULAS SIN AÑO RECONOCIBLE: {total_no_year} (de un total de {len(movies_clean)})")
print("-" * 60)

print("Muestra de títulos detectados como CASOS LÍMITE:")
if total_no_year > 0:
    display(no_year_movies[['movieId', 'title']].head(10))

print("\n" + "="*70 + "\n")

print("VISTA PREVIA DEL DATAFRAME FINAL ('movies_clean')")
display(movies_clean.head(10))
```

```python
# --- RAUL ---
# --- Parte 2: TMDB GET /movie/{id} → overview y homepage  ---
# Requiere: `movies` y `links` cargados. pip install requests

import requests
import os
import json
from dotenv import load_dotenv
load_dotenv()
TMDB_API_KEY = os.getenv("TMDB_API_KEY")

params = {"api_key": TMDB_API_KEY}
    

def fetch_movie_details(tmdb_id) -> list[str, str]:
    try:
        url = f"https://api.themoviedb.org/3/movie/{tmdb_id}"
        resp = requests.get(url, params=params, timeout=10)
        resp.raise_for_status()
        
        data = dict(resp.json())
        print(json.dumps(data, ensure_ascii=False, indent=2))
        return [(data.get("homepage") or "").strip(), (data.get("overview") or "").strip()]
    except requests.exceptions.RequestException as e:
        print(f"Excepción capturada: {e}")
        raise e

# RAUL: Si no hay link, no quiero que aparezca, y si no hay título tampoco, por tanto inner.
movies10 = pd.merge(df_movies, df_links, how="inner", on="movieId").head(10)
display(movies10)
print(type(movies10))

print(fetch_movie_details(100))

lista = movies10["tmdbId"].apply(fetch_movie_details)
# RAUL: Esto me castea la lista a series, porque sino no funciona.
movies10[["homepage","overview"]] = lista.apply(pd.Series)

display(movies10)
```

```python
# --- RAUL ---
# --- Parte 3: overview → overview_es con Gemini  ---
# Requiere: `movies10` de la Parte 2. pip install google-genai

from google import genai
import time

MODEL = "gemini-2.5-flash"
client = genai.Client()
retries = 3

def summarize_overview_es(overview, title=""):
    last_err = None
    for attempt in range(retries + 1):
        try: 
            response = client.models.generate_content(
                model=MODEL,
                contents=f"Devuelveme un resumen en español del siguiente texto: {overview} que corresponde a la película {title}. \
                    Debes usar como máximo dos frases.",)
            return response.text or ""
        except Exception as e:
            last_err = e
            if attempt >= retries:
                break
            time.sleep(2 ** attempt)
    print(f"No se pudo obtener respuesta tras reintentos. Último error: {last_err}")
    return ""

for overview, title in zip(movies10["overview"], movies10["title"]):
    movies10.loc[movies10["title"] == title, ["overview_es"]] = summarize_overview_es(overview, title)
    # RAUL: Parece que Gemini se satura. Voy a esperar 10 segundos entre intentos.
    time.sleep(30)

# movies10.loc[movies10["title"] == movies10[0], ["overview_es"]] = summarize_overview_es(movies10[0][] title)
display(movies10.head(3))
```
